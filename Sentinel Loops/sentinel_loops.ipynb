In this notebook, you will implement a sentinel loop to avoid your system from keep on running when a solution has already been found within a process that only requires a single solution, while avoiding other alternatives such as `break` statements.

The following code is already able to find the lowest number in the interval [1, 1000) that is divisible by all numbers within a given list. This number is also commonly referred to as the least common multiple or the lowest common multiple. This solution is implemented using a for loop. Modify the two lines of code that are indicated below to provide a new solution that uses a sentinel loop instead.

```python
def lcm_lt1000(numbers):
    
    result = None
    
    for i in range(1, 1000):  # Modify this line of code.
        
        if sum([i % number for number in numbers]) == 0:
            result = i    
            print("LCM Result is {0}".format(result))
            break  # Modify this line of code.
    
    return result
```

You may use the following assertion clauses to check if your code is working properly. If the code is able to reach the print clause, everything worked as expected. Otherwise, an exception will be raised before that. You can also use these sentences to check the expected format returned by the function.

```python
assert lcm_lt1000([3, 4, 6]) == 12
assert lcm_lt1000([2, 4, 5]) == 20
assert lcm_lt1000([2, 3, 5]) == 30
assert lcm_lt1000([2, 5, 7]) == 70
print('Code tested successfully.')
```
